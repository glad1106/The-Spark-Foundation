```python
txt["publish_date"] = txt["publish_date"].astype(str)
```

```python
txt["publish_date"] = txt["publish_date"].apply(lambda x: x[0:4]+"-"+x[4:6]+"-"+x[6:])
```

```python
txt["publish_date"] = txt["publish_date"].astype("datetime64")
```

```python
txt.info()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
mod = SentimentIntensityAnalyzer()
```

```python
mod.polarity_scores(txt.iloc[50,-1])
```

```python
txt["Polarity"] = txt["headline_text"].apply(lambda x: mod.polarity_scores(x))
```

```python
txt
```

```python
txt["Compound_Score"] = txt["Polarity"].apply(lambda x: x["compound"])
```

```python
txt
```

```python
tx = txt[["publish_date","Compound_Score"]]
tx
```

```python
tx = tx.rename(columns ={"publish_date":"Date"})
```

```python
tx
```

```python
df = pd.merge(stock,tx, on= "Date", how = "left")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
np.column_stack((y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```
