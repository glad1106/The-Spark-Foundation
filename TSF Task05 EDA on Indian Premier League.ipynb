# Visualizing The Data

```python
x=df_final['inning'].unique()
x
```

```python
l=[]
for i in df_final['inning'].unique():
    print(df_final[df_final['inning']==i]['total_runs'].sum())
    l.append(df_final[df_final['inning']==i]['total_runs'].sum())
```

```python
plt.figure(figsize=(24,8))
plt.plot(x,l,color='green',marker='o',linestyle='dashed',markersize=12)
plt.xlabel("Innings Number")
plt.ylabel("Total Runs Made")
plt.show()
```

# Total Runs Made In Each City

```python
plt.figure(figsize=(16,16))
sns.countplot(x=df_final['city'],data=df_final)
plt.xlabel("City Name")
plt.ylabel("Total Runs")
plt.xticks(rotation = 90)
plt.show
```

# Total Runs Scored By Team

```python
plt.figure(figsize=(16,16))
sns.countplot(x=df_final['winner'],data=df_final)
plt.title("Team Scored Total Runs")
plt.xlabel("Team Name")
plt.ylabel("Total Runs")
plt.xticks(rotation = 90)
plt.show
```

# Number Of Unique Stadiums Teams Played On

```python
plt.figure(figsize=(16,16))
df_final.groupby('batting_team')['venue'].nunique().plot(kind='bar',fontsize=22)
plt.xlabel("Team Name")
plt.ylabel("Played Match On Stadium")
```

# Team Played In Each Season

```python
plt.figure(figsize = (16,16))
df_final.groupby('season')['team1'].nunique().plot(kind='bar')
plt.title("Team Played In Each Season")
plt.xlabel("Season",fontsize=30)
plt.ylabel("Number Of Team",fontsize=15)
```

# Number Of Times A Team Won A Match

# Season Wise Winning Teams

```python
x=df_final[['season','winner']]
team_name = {}
for i in sorted (x.season.unique()):
    team_name[i]=x[x.season==i]['winner'].tail(1).values[0]
winners=pd.Series(team_name)
winners=pd.DataFrame(winners,columns=['team_name'])
print("Winners Are:",winners)
```

```python
plt.figure(figsize=(8,8))
winners['team_name'].value_counts().plot(kind='bar')
plt.title("Winners Of IPL",fontsize=30)
plt.xlabel("Team Name",fontsize=20)
plt.ylabel("No. Of Teams Won",fontsize=20)
plt.show
```
